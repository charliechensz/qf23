```python
# # 导入相关的库
# import tushare as ts      
# import pandas as pd   
# import matplotlib.pyplot as plt

# # 设置token，即接口密匙, 获取方式及相关权限详情见Tushare官网。
# token = 'Your token'     
# pro = ts.pro_api(token)

# 获取日行情的接口。
df = pro.daily(ts_code='002371.SZ', start_date='20200101', end_date='20200116')   
df['trade_date'] = pd.to_datetime(df['trade_date'])  
df.set_index(['trade_date'], inplace=True)

# 倒序，因为Tushare的数据最近日期默认显示在DataFrame上方位置。倒序方便后作图。
df = df.sort_index(ascending=True)  
# 取2020年的数据
Close = df.close['2020']   
Open = df.open['2020']
Vol = df.vol['2020']
```

```python
# plt.rcParams['font.sans-serif'] = ['SimHei']   # 显示中文设置

ax1 = plt.subplot(211)  #2行1列
ax1.plot(Close, color='k')  
ax1.set_ylabel('收盘价')
ax1.set_title('北方华创2020年收盘价曲线图')
# ax1.set_ylim(150, 200)   

ax2 = plt.subplot(212)
left1 = Vol.index[Close>=Open]
hight1 = Vol[left1]
ax2.bar(left1, hight1, color='r')

left2 = Vol.index[Close<Open]
hight2 = Vol[left2]
ax2.bar(left2, hight2, color='g')

ax2.set_ylabel('成交量')
ax2.set_title('北方华创2020年成交量柱状图')
```

```python
# 提取出最高价和最低价
High = df.high['2020']
Low = df.low['2020']

# plt.rcParams['font.sans-serif'] = ['SimHei']   # 显示中文设置

# 创建Axes对象
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.25, 0.9, 0.9])

# 绘制四条曲线
# ax.plot(Open, '--*', label='开盘价')
ax1.plot(Close, label='收盘价')

# ax.plot(High, '--o', label='最高价')
# ax.plot(Low, '-.>', label='最低价')

# ax.bar(df.index, Close, color='g')
# 在axes对象中增加文本对象
ax1.set_title('北方华创2020年价格图')
ax1.set_ylabel('价格')
ax1.legend(loc='best')
ax1.scatter(df.index, Open, c='g',s=10, alpha=0.3, cmap='viridis')

```
