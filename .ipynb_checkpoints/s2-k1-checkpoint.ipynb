```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from matplotlib.font_manager import FontProperties

import mplfinance as mpf
# from matplotlib.dates import date2num
# from cycler import cycler
# font = FontProperties(fname="SimHei.ttf", size=14)
```

```python
# test2 : link to pandas 
   
df0 = pd.read_csv('601318.csv', index_col='trade_date', parse_dates=True)
df0.loc[:, "volume"] = df0["amount"]
df = df0[['open', 'close', 'high', 'low', 'volume']].sort_index(ascending=True)
df
# df.plot()
# plt.show()
```

```python
%time
daily = df[['open',  'high', 'close','low', 'volume']].loc['2021-07-01':,:]
df
```

```python
# 设置mplfinance的蜡烛颜色，up为阳线颜色，down为阴线颜色
my_color = mpf.make_marketcolors(up='r',
                                 down='g',
                                 edge='inherit',
                                 wick='inherit',
                                 volume='inherit')
# 设置图表的背景色
my_style = mpf.make_mpf_style(marketcolors=my_color,
                            figcolor='(0.82, 0.83, 0.85)',
                            gridcolor='(0.82, 0.83, 0.85)')

mpf.plot(daily, style=my_style, type='candle',
        mav=(5,10,20,48), volume=True, figratio=(16, 9), figscale=1.2)

# Ver 1.0
# kwargs = dict(type='candle', mav=(5,10,20,48), volume=True, figratio=(16,9), figscale=1.3)
# mpf.plot(daily, **kwargs, style='charles')
# mpf.plot(daily, type='candle', mav=(5, 10, 48), volume=True, show_nontrading=False)
# mpf.plot(daily, type='line')
# mpf.plot(daily, type='renko')
plt.show()
```
