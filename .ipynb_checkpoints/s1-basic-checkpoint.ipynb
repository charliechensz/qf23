```python
# 导入数据
# df = pd.read_csv('600519.csv', index='trade_date',    parse_dates='trade_date')
df = pd.read_csv('600519.csv', index_col="trade_date", parse_dates=True )[['open','close','high','low']]
# df
```

```python
# 上涨超过3%的day
# df[(df['close']-df['open'])/df['open']>0.03].index
df[(df['close']-df['open'])/df['open']>0.03]
```

```python
# 低开2%的day， 遍历繁琐，用
df[ ( df['open'] - df['close'].shift(1) ) / df['close'].shift(1) > 0.02 ]
# shiftBracket Pair Colorizer 2： 括号颜色区分
```

```python
#每月1号买入，年底12月31号卖出。收益多少？
df = df['2001-09-01':'2017-11-30']   ## 剔除首尾无用数据
df.resample('M')
# df.resample('MS').first()
df
```

```python
# 每月第一天的数据
# df.resample('M').sum()
# df.resample('M').mean()  ## 平均值
# df.resample('M').first()  ## 每月第1天数据， index显示最后1天
# df.resample('M').last()  ## 每月最后1天数据， index预设显示最后1天
df_monthly = df.resample('MS').first()  ## 每月第1天， index显示第1天
df_monthly
```
