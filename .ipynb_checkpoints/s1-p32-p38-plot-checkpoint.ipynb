```python
# test 3 :
x = np.linspace(-100, 100, 10000)
y1 = x.copy()
y2 = x**2
y3 = x**3 + 5*x**2 + 2*x + 1 
plt.plot(x, y1, color='red', label='y=x')
plt.plot(x, y2, color='blue', label='y=x**2')
plt.plot(x, y3, color='green', label='y=x**3')
plt.title("Title 3 ")
plt.legend()
plt.xlabel("Number")
# plt.ylabel("Height")
plt.xlim(-10, 10)    # X轴输出范围
plt.ylim(-100, 100)   # Y轴输出范围
# plt.xticks( np.arange(0, 10, 3), ['A', "B", "C", 'D'])   # 刻度
plt.xticks( np.arange(-100 , 100, 20))   # 刻度
plt.show()
```

```python
# test4:  画布， canvos
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)  # 分成4格，在左上角
ax1 = plt.plot([1,2,3,4], [1,2,3,4])
ax2 = fig.add_subplot(2, 2, 2)  # 分成4格，在左上角
ax3 = fig.add_subplot(2, 2, 3)  # 分成4格，在左上角
ax3 = plt.plot([1,2,3,4], [4, 3, 2, 1])
ax4 = fig.add_subplot(2, 2, 4)  # 分成4格，在左上角
plt.show()
```

```python
# test 5: 更多显示类型, 柱状图

# plt.bar([0,1,2,8], [1,2,3, 4])
data = [32, 48, 21, 100]
labels = ['Jan', 'Feb', 'Mar', 'Apr']

plt.bar( np.arange(len(data)), data )
plt.xticks( np.arange( len(data) ), labels )
plt.show()
```
