    mplfinance：['open','high','low','close','pre_close','volume', 'change', 'pct_change', 'value']
- 批量获取目标数据 到 csv 格式



```python
%matplotlib widget
import numpy as np
import pandas as pd
import random 
import matplotlib.pyplot as plt
import mplfinance as mpf

# from numba import jit

import tushare as ts
ts.set_token('11137efdbeac800606d677871a55b3fd5aef79890c59867a8f34d03e')
pro = ts.pro_api()
```

```python
# a = '000001'
# a.startswith('1')
ts_list = pd.read_csv( 'data/stock_list.csv')
# ts_list = pd.read_csv( 'data/stock_list_first.csv')
# ts_list = pd.read_csv( 'data/stock_list_all.csv')
ts_list

def cal_upper(df):
    return ( df['high'] - max( df['open'], df['close'] ) ) / df['pre_close'] * 100
def cal_down(df):
    return ( min( df['open'], df['close'] ) - df['low']  ) / df['pre_close'] * 100
def cal_range(df):
    return ( df['high'] - df['low']  ) / df['pre_close'] * 100
def cal_st_high(df):
    return df['open'] if df['open'] >= df['close'] else df['close']
def cal_st_low(df):
    return df['open'] if df['open'] < df['close'] else df['close']
def cal_st_range(df):
    return abs( df['open'] - df['close']  ) / df['pre_close'] * 100

    # if df['close'] >= df['open']:
    #     return df['weight']*0.7
    # #成年男女
    # if one_row['sex']=='M':
    #     return one_row['weight']*0.6
    # else:
    #     return one_row['weight']*0.5
```

```python
%%time

start_date = '2019-01-01'
end_date = '2022-12-31'

for i in range( ts_list.shape[0]  ):
    print('收集数据：', ts_list.index[i], ts_list['ts_code'][i])
    # 国内股市：可显示复权后和复权前信息, 现在用前复权价格
    df = ts.pro_bar(ts_code=ts_list['ts_code'][i], asset='E', adj='qfq', start_date=start_date, end_date='')
    # 字段改名
    # df = df.sort_index(ascending=True)
    df.rename( columns={'vol':'volume','pct_chg':'pct_change', 'amount':'value'}, inplace=True  )
    
    # 实体部分细化
    df['pct_upper'] = df.apply(cal_upper, axis=1)   # 上引线百分比
    df['pct_down']  = df.apply(cal_down,  axis=1)   # 下引线百分比
    df['pct_range'] = df.apply(cal_range, axis=1)   # 实体部分百分比
    df['st_high']       = df.apply(cal_st_high,  axis=1)   # 实体高点价格
    df['st_low']        = df.apply(cal_st_low,   axis=1)   # 实体低点价格
    df['pct_st_range']  = df.apply(cal_st_range, axis=1)   # 实体部分百分比

    # 星线定义
        
    ts_file = 'tushare/' + ts_list['ts_code'][i] +'.ts.csv'
    df.to_csv( ts_file )
    
```
