## S2, DB结构

    第一个字段为日期，没有字段名 
    open: 当日开盘价 
    high, 当日最高价
     low, 当日最低价 
    close, 当日收盘价 
    volume, 交易量（手） 
    MA5, 收盘价的5日移动平均线 
    MA10, 收盘价的10日移动平均线 
    MA20, 收盘价的20日移动平均线 
    MA60, 收盘价的60日移动平均线 
    change, 当日收盘价相对上个收盘价的变化值 
    pct_change, 当日收盘价相对上个收盘价的涨跌幅百分比 
    value, 交易额（亿元） 
    upper_lim, 涨停价格 
    lower_lim, 跌停价格 

    last_close, 昨日收盘价 
    average, 平均价 
    volrate, 交易量（手） 

    dema, DEMA指标值 
    macd-m, MACD指标-M线， macd 
    macd-s, MACD指标-S线， singal
    macd-h, MACD指标-红绿柱, history
    rsi, RSI指标 
    bb-u, 布林带线上轨 
    bb-m, 布林带线中轨
    bb-l, 布林带线下轨

- dea,  macd-s, signal = 9 天 MACD的EMA

```python
# 计算macd
# 输入df_name,算法参数。其中ewm函数算法快300-500倍
def cal_macd(dfIn=df, macd_way=2, ema_fast=12, ema_slow=26, dea_day=9):
    dfIn['ema_fast'] = dfIn['close'].ewm(span=ema_fast,     adjust=False).mean()
    dfIn['ema_slow'] = dfIn['close'].ewm(span=ema_slow,     adjust=False).mean()
    dfIn['diff'] = dfIn['ema_fast'] - dfIn['ema_slow']
    dfIn['dea'] = dfIn['diff'].ewm(span=9, adjust=False).mean()
    dfIn['bar'] = (dfIn['diff'] - dfIn['dea']) * 2
    return dfIn

@jit(nopython=True)
def jit_cal_macd(dfIn=df, macd_way=2, ema_fast=12, ema_slow=26, dea_day=9):
    dfIn['ema_fast'] = dfIn['close'].ewm(span=ema_fast,     adjust=False).mean()
    dfIn['ema_slow'] = dfIn['close'].ewm(span=ema_slow,     adjust=False).mean()
    dfIn['diff'] = dfIn['ema_fast'] - dfIn['ema_slow']
    dfIn['dea'] = dfIn['diff'].ewm(span=9, adjust=False).mean()
    dfIn['bar'] = (dfIn['diff'] - dfIn['dea']) * 2
    return dfIn

# 计算MA
def cal_ma_v0():
    df['MA5']  = df['close'].rolling(5).mean()
    df['MA10'] = df['close'].rolling(10).mean()
    df['MA20'] = df['close'].rolling(20).mean()
    df['MA30'] = df['close'].rolling(30).mean()
    df['MA48'] = df['close'].rolling(48).mean()
    
def cal_ma_v1(dfIn=df):
    dfIn['MA5'] = dfIn['close'].rolling(5).mean()
    dfIn['MA10'] = dfIn['close'].rolling(10).mean()
    dfIn['MA20'] = dfIn['close'].rolling(20).mean()
    dfIn['MA30'] = dfIn['close'].rolling(30).mean()
    dfIn['MA48'] = dfIn['close'].rolling(48).mean()
    return dfIn

@jit(nopython=True)
def jit_cal_ma(dfIn=df):
    df['MA5'] = df['close'].rolling(5).mean()
    df['MA10'] = df['close'].rolling(10).mean()
    df['MA20'] = df['close'].rolling(20).mean()
    df['MA30'] = df['close'].rolling(30).mean()
    df['MA48'] = df['close'].rolling(48).mean()
    # return df.dropna()
    # dfIn['MA5'] = dfIn['close'].rolling(5).mean()
    # dfIn['MA10'] = dfIn['close'].rolling(10).mean()
    # dfIn['MA20'] = dfIn['close'].rolling(20).mean()
    # dfIn['MA30'] = dfIn['close'].rolling(30).mean()
    # dfIn['MA48'] = dfIn['close'].rolling(48).mean()
    # # df['MA60'] = np.nan
    # # df['MA60'] = df['close'].rolling(60).mean()
    # return dfIn.dropna()
```

```python
%%time

# ts_list = pd.read_csv( 'data/stock_list_first.csv')
ts_list = pd.read_csv( 'data/stock_list_first.csv')
for i in range( ts_list.shape[0]  ):
    if ts_list['ts_code'][i].startswith('8'):
        break
    else:
        print('计算：', ts_list['ts_code'][i] )
    
    ts_file = 'tushare/' + ts_list['ts_code'][i] +'.ts.csv'
    df = pd.read_csv( ts_file, index_col='trade_date', parse_dates=True).sort_index(ascending=True)
    # df = df.sort_index(ascending=True)
    df.rename( columns={'vol':'volume','pct_chg':'pct_change', 'amount':'value'}, inplace=True  )
 
    # 计算MACD
    # way1 : charlie's coding
    # df['ema_fast']=0.0
    # df['ema_slow']=0.0
    # df['diff']=0.0
    # df['dea']=0.0
    # df['bar']=0.0
    # df = cal_macd(df, 2, 12, 26, 9)
    # way2 : ta-lib
    close = df.close.values  # 为 MACD 数据初始化close
    df['diff'], df['dea'], df['bar'] = ta.MACD(close, fastperiod=12, slowperiod=26, signalperiod=9)
    
    # 计算MA
    # way1: my coding
    # df['MA5'] = np.nan
    # df['MA10'] = np.nan
    # df['MA20'] = np.nan
    # df['MA30'] = np.nan
    # df['MA48'] = np.nan
    # df = cal_ma_v1(df)
    # way 2: ta-lib
    df.insert( df.shape[1], 'MA5', ta.MA(df['close'], 5) )
    df.insert( df.shape[1], 'MA10', ta.MA(df['close'], 10) )
    df.insert( df.shape[1], 'MA20', ta.MA(df['close'], 20) )
    df.insert( df.shape[1], 'MA30', ta.MA(df['close'], 30) )
    df.insert( df.shape[1], 'MA48', ta.MA(df['close'], 48) )
    df.insert( df.shape[1], 'MA60', ta.MA(df['close'], 60) )
    df.insert( df.shape[1], 'MA125', ta.MA(df['close'], 125) )
    
    # 计算BOLL, 
    df['bb-u'], df['bb-m'], df['bb-l'] = ta.BBANDS( close,timeperiod=20,nbdevup=2,nbdevdn=2,matype=0 )
    # df['upper'], df['middle'], df['lower'] = ta.BBANDS( close,timeperiod=20,nbdevup=2,nbdevdn=2,matype=0 )
    # 空值填充为0
    df = df.fillna(value=0)

    # 输出 csv
    ts_cal_file = 'cal/' + ts_list['ts_code'][i] +'.cal.csv'
    df.to_csv(ts_cal_file)
df
    

# df['upper'], df['middle'], df['lower'] = ta.BBANDS( df.close.values,timeperiod=20,nbdevup=2,nbdevdn=2,matype=0 )
# MA60  = ta.MA(df['close'], 60)
# MA120 = ta.MA(df['close'], 120)

# 插入和删除行，很有用

# df.insert( df.shape[1], 'MA250', ta.MA(df['close'], 250) )
# df.drop( labels=['MA125'], axis=1, inplace=True)

```
